```python
import pandas as pd
from nltk.tokenize import word_tokenize
import string
import re
import nltk
nltk.download('punkt_tab')

punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

df = pd.read_csv("datasetberita.csv", header=None, sep='\r\n', engine='python', names=['berita'])

dataset = pd.read_csv('datasetberita.csv')
dataset.head(20)
```

```python
#lowercase
df['lowercase'] = df['berita'].apply(lambda x: (x.lower()))

print (df['lowercase'])


```

```python
#generate tokens
df['tokens'] = df['lowercase'].apply(lambda x: word_tokenize(x))

print(df['tokens'])
```

```python
df['lowercase'] = df['lowercase'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))


print(df['lowercase'])
```

```python
df['lowercase'] = df['lowercase'].apply(lambda x: (re.sub(r'[\w\s]', '', x)))
print (df['lowercase'])
```
